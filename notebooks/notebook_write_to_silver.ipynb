```python
numeric_column = [
    "trip_distance",
    "fare_amount",
    "total_amount"
]
```

```python
import pyspark.sql.functions as f
for column in numeric_column:
    less_Q1 = 'less_Q1_{}'.format(column)
    more_Q3 = 'more_Q3_{}'.format(column)
    Q1 = 'Q1_{}'.format(column)
    Q3 = 'Q3_{}'.format(column)

    #Q1 : First Quartile., Q3 : Third Quartile
    Q1 = df_clean.approxQuantile(column, [0.25], 0.01)
    Q3 = df_clean.approxQuantile(column, [0.75], 0.01)

    #IQR : Inter Quantile Range
    # We need to define the index [0], as Q1 & Q3 are a set of lists., to perform a mathematical operation
    # Q1 & Q3 are defined separately so as to have a clear indication on First Quantile & 3rd Quantile
    IQR = Q3[0] - Q1[0]
    
    less_Q1 = Q1[0] - 1.5*IQR
    more_Q3 = Q3[0] + 1.5*IQR
    isOutlierCol = 'is_outlier_{}'.format(column)
    df_clean = df_clean.withColumn(isOutlierCol,f.when((df_clean[column] > more_Q3) | (df_clean[column] < less_Q1),1).otherwise(0))
# Selecting the specific columns which we added above, to check if there are any outliers
selected_columns = [column for column in df_clean.columns if column.startswith("is_outlier")]

#Adding all the outlier columns into a new column "total_outliers, to see the total number of outliers"
df_clean = df_clean.withColumn('total_outliers', sum(df_clean[column] for column in selected_columns))
display(df_clean)
# df_clean = df_clean.drop(*[column for column in df_clean.columns if column.startswith("is_outlier")])
```

```python
import requests
import time
from pyspark.sql import functions as F
df_aq_meas = df_aq.select(
    F.col("id"), 
    F.col("`sensors.id`").alias("sensor_id"), 
    F.col("`sensors.parameter.name`").alias("parameter"))
print(df_aq_meas)
```

```python
from datetime import datetime, timedelta

months = []
current = datetime(2022, 1, 1)
while current <= datetime(2025, 12, 1):
    next_month = datetime(current.year + (current.month == 12), 
                         (current.month % 12) + 1, 1)
    months.append((
        current.strftime("%Y-%m-%dT00:00:00Z"),
        (next_month - timedelta(seconds=1)).strftime("%Y-%m-%dT23:59:59Z")
    ))
    current = next_month

print(f"Total count of months: {len(months)}")
print(months[:3])
```

```python
import requests
import time
from datetime import datetime, timedelta

sensors_list = df_aq_meas.collect()
all_measurements = []

for s in sensors_list:
    location_id = s[0]
    sensor_id = s[1]
    parameter = s[2]
    
    for date_from, date_to in months:
        page = 1
        retries = 0
        
        while True:
            url = f"https://api.openaq.org/v3/sensors/{sensor_id}/measurements"
            params = {
                "datetime_from": date_from,
                "datetime_to": date_to,
                "limit": 1000,
                "page": page
            }
            headers = {"X-API-Key": "30e0ed845a7c170412423e53cccdcc77d8168a2c51b14af06c667c02a4b44601"}
            
            response = requests.get(url, params=params, headers=headers, timeout=30)
            
            if response.status_code in [408, 500]:
                retries += 1
                if retries >= 3:
                    print(f"Miss sensor {sensor_id} {date_from[:7]}")
                    break
                time.sleep(5)
                continue
                
            if response.status_code == 429:
                time.sleep(10)
                continue
                
            if response.status_code != 200:
                print(f"Miss sensor {sensor_id}: status {response.status_code}")
                break
                
            data = response.json()
            results = data.get("results", [])
            
            if not results:
                break
                
            for r in results:
                all_measurements.append({
                    "location_id": location_id,
                    "sensor_id": sensor_id,
                    "parameter": parameter,
                    "value": r["value"],
                    "datetime_from_utc": r["period"]["datetimeFrom"]["utc"],
                    "datetime_to_utc": r["period"]["datetimeTo"]["utc"]
                })
            
            print(f"Sensor {sensor_id} {date_from[:7]}: {len(results)} rows")
            
            if len(results) < 1000:
                break
                
            page += 1
            time.sleep(1)

print(f"All measurements: {len(all_measurements)}")
df_measurements = spark.createDataFrame(all_measurements)
df_measurements.write.format("delta").mode("overwrite").saveAsTable("Bronze.dbo.raw_openaq_measurements")
print("Saved in Bronze!")
```
